```python
guitar = NSGData.from_wav("./samples/guitar.wav", 
GaborSystemParams(
    scale=Oct,
    n_b=60,
    fmin=30,
    fmax=22000,
    fs=44100,
    Ls=44100,
    reducedform=0
))

guitar.to_tiff("./images/guitar.tiff")

guitar_r = NSGData.from_tiff("./images/guitar.tiff")


from torch import allclose
allclose(guitar.s, guitar_r.s, atol=1e-7)
```
